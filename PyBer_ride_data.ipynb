```python
# Add dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load in Csv
pyder_rider_df=pd.read_csv("C:/Users/avnit/PyBer_Analysis/Resources/PyBer_ride_data.csv")
pyder_rider_df
```

```python
# Set x-axis and tick location:
x_axis= np.arange(len(pyder_rider_df))
tick_location=[value for value in x_axis]
pyder_rider_df.plot(x="Month",y="Avg. Fare ($USD)")
plt.xticks(tick_location,pyder_rider_df["Month"])
plt.show()
```

```python
pyder_rider_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```
