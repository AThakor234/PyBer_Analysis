```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legends.
plt.plot(x_axis,y_axis,marker = "*", color = "blue",linewidth = 2,label = "Boston")

# Create labels for x and y axis

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45:
plt.ylim(0,45)
    
# Create  a title.
plt.title("Pyber Fare by Month")

# Add the legend

plt.legend()
```

```python
# Bar Chart
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis,y_axis, color="green", label="Boston")

# Create label for x and y axis

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create a horizontal chart
plt.barh(x_axis,y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis ,color="magenta",label = "Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("(Fare(S))")
plt.legend()
```

```python
# Create the plot with ax.plot()
fig,ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
plt.scatter(y_axis,x_axis,color="red",label="Chicago")
plt.legend()
plt.title("PyBer fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Using Object-Oriented

figs,ax = plt.subplots()
ax.scatter(y_axis,x_axis, c="skyblue", s = [i*5 for i in y_axis], alpha = 0.80, edgecolor = "black",linewidths=2,label="Boston")
ax.legend()
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
# Pie Charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
explode_values =(0,0,0,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, labels=x_axis, explode = explode_values, autopct ='%.1f%%',colors=colors)
plt.subplots(figsize=(8,8))

```

```python
# Oject-Oriented Method[Pie]
fig,ax=plt.subplots()
explode_values =(0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values,shadow=True,colors=colors,startangle=90,counterclock=False)
plt.subplots(figsize=(8,8))
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation

stdev= statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
# Using Object-Oriented Method

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Adding errorbar to barchart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
