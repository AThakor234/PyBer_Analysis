```python
# Combining two data sets
pyber_data=pd.merge(ride_data,city_data,how="left",on=["city","city"])
pyber_data.head()
```

```python
# Create the Urban city Data Frame
urban_city= pyber_data[pyber_data["type"]=="Urban"]
urban_city
```

```python
# Create the Suburban Data Frame
suburban_city = pyber_data[pyber_data["type"]=="Suburban"]
suburban_city
```

```python
# Create Rural Data Frame
rural_city= pyber_data[pyber_data["type"]=="Rural"]
rural_city
```

```python
# Get the number of rides for urban cities.
urban_ride_count=urban_city.groupby(["city"]).count()["ride_id"]
urban_ride_count
```

```python
# Get the number of rides for sunurban cities
suburban_ride_count=suburban_city.groupby(["city"]).count()["ride_id"]
suburban_ride_count
```

```python
# Get the number of rides in rural cities
rural_ride_count=rural_city.groupby(["city"]).count()["ride_id"]
rural_ride_count
```

```python
# Get the average fare for Urban cities
urban_fare = urban_city.groupby(["city"]).mean()["fare"]
urban_fare
```

```python
# Get the average fare for Suburban cities
suburban_fare = suburban_city.groupby(["city"]).mean()["fare"]
suburban_fare
```

```python
# Get the average fare for Rural cities
rural_fare= rural_city.groupby(["city"]).mean()["fare"]
rural_fare
```

```python
# Get the average number of drivers for each urban city.
urban_driver_count = urban_city.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head()
```

```python
# Get the average number of drivers for Suburban city
suburban_driver_count= suburban_city.groupby(["city"]).mean()["driver_count"]
suburban_driver_count
```

```python
# get the average number of drivers for Rural
rural_driver_count= rural_city.groupby(["city"]).mean()["driver_count"]
rural_driver_count
```

```python
# Scatter Plot for Urban Cities
plt.scatter(urban_ride_count,urban_fare,
            s=10*urban_driver_count,
            color="coral",
            edgecolor="black",
            linewidth=1,
            alpha = 0.8,
            label="Urban")
plt.title("PyBer Ride-Sharing Data(2019)")
plt.ylabel("Average Fare($)")
plt.xlabel("Total Number of Rides(Per City)")
plt.legend()
plt.grid(True)
```

```python
# Creating Scatter Plot for Suburban cities
plt.scatter(suburban_ride_count,suburban_fare,
           color="skyblue",
           s=10*suburban_driver_count,
           alpha=0.80,
           edgecolor="black",linewidth=1,
           label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Creating scatter chart for Rural
plt.scatter(rural_driver_count,
      rural_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")


plt.grid(True)

plt.legend()
```

```python
# Scatter the plot in one chart

plt.subplots(figsize=(10,6))

plt.scatter(urban_ride_count,urban_fare,
            s=10*urban_driver_count,
            color="coral",
            edgecolor="black",
            linewidth=1,
            alpha = 0.8,
            label="Urban")



plt.scatter(suburban_ride_count,suburban_fare,
           color="skyblue",
           s=10*suburban_driver_count,
           alpha=0.80,
           edgecolor="black",linewidth=1,
           label="Suburban")



plt.scatter(rural_driver_count,
      rural_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")


plt.title("PyBer Ride-Sharing Data (2019)",fontsize=20)
plt.ylabel("Average Fare ($)",fontsize=12)
plt.xlabel("Total Number of Rides (Per City)",fontsize=12)
plt.legend()
plt.grid(True)
plt.xlim(0,40)

# Create a legend
lgnd = plt.legend(fontsize="12",mode="expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")

plt.show()





```
